```python
import pandas as pd
import os
import datetime
```

```python
file1 = "Resources/JC-202201-citibike-tripdata.csv"
file2 = "Resources/JC-202202-citibike-tripdata.csv"
file3 = "Resources/JC-202203-citibike-tripdata.csv"
file4 = "Resources/JC-202204-citibike-tripdata.csv"
file5 = "Resources/JC-202205-citibike-tripdata.csv"
file6 = "Resources/JC-202206-citibike-tripdata.csv"

file_list = [file1,file2,file3,file4,file5]
df_list = []

for file in file_list:
    df = pd.read_csv(file)
    df_list.append(df)

print(df_list)
```

```python
df_merged = pd.concat(df_list)
df_merged.head()
```
